### Daily Task Manager Web Application Logic with Front End

```python
## Logic -> logic.py

tasks = []

def add_task(title, priority="Medium"):
    task = {
        "id": len(tasks) + 1,
        "title": title,
        "priority": priority,
        "completed": False
    }
    tasks.append(task)
    return task

def toggle_task(task_id):
    for task in tasks:
        if task["id"] == task_id:
            task["completed"] = not task["completed"]
            return task
    return None

# Test the logic
add_task("Pay Car Insurance", "High")
print(tasks)

```

### Logic with Front End

```python
import IPython
from google.colab import output

tasks = []

def add_task_backend(title):
    """The Python logic that would eventually save to a real DB."""
    if title:
        task = {"id": len(tasks) + 1, "title": title, "completed": False}
        tasks.append(task)
        print(f"Python received: {title}. Total tasks: {len(tasks)}")
        return tasks
    return tasks

# Register the function so JavaScript can "see" it
output.register_callback('notebook.add_task', add_task_backend)
```

### Frontend (HTML + CSS + JS) Integration

```python
html_prototype = """
<!DOCTYPE html>
<html>
<head>
  <script src="https://cdn.tailwindcss.com"></script>
  <link href="https://fonts.googleapis.com/css2?family=Inter:wght@400;600&display=swap" rel="stylesheet">
  <style> body { font-family: 'Inter', sans-serif; } </style>
</head>
<body class="bg-gray-100 p-10">

  <div class="max-w-md mx-auto bg-white rounded-2xl shadow-lg p-6">
    <h2 class="text-xl font-bold mb-4 text-gray-800">Prototyping: Daily Tasks</h2>

    <div class="flex gap-2 mb-6">
      <input id="taskInput" type="text" placeholder="What needs doing?"
             class="flex-1 p-2 border rounded-lg outline-none focus:ring-2 focus:ring-blue-400">
      <button onclick="sendToPython()"
              class="bg-blue-600 text-white px-4 py-2 rounded-lg hover:bg-blue-700 transition">
        Add
      </button>
    </div>

    <ul id="taskList" class="space-y-2">
      </ul>
  </div>

  <script>
    async function sendToPython() {
      const input = document.getElementById('taskInput');
      const val = input.value;

      if (!val) return;

      // BRIDGE: Call the Python function registered above
      // It returns a Promise with the updated task list
      const result = await google.colab.kernel.invoke_method('notebook.add_task', [val], {});

      updateUI(result.data['application/json']);
      input.value = '';
    }

    function updateUI(tasks) {
      const list = document.getElementById('taskList');
      list.innerHTML = tasks.map(t => `
        <li class="p-3 bg-gray-50 rounded-lg border border-gray-200 flex items-center justify-between">
          <span class="text-gray-700 font-medium">${t.title}</span>
          <span class="text-xs text-gray-400">#${t.id}</span>
        </li>
      `).join('');
    }
  </script>
</body>
</html>
"""

IPython.display.HTML(html_prototype)
```
